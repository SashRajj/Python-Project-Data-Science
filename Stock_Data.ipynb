# Question 1:  Extracting Tesla Stock Data Using yfinance

```python
import yfinance as yf
import pandas as pd

# Download Tesla data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Reset index
tesla_data.reset_index(inplace=True)

# Display first 5 rows
print(tesla_data.head())
```

# Question 2: Extracting Tesla Revenue Data Using Webscraping

```python
# Get quarterly revenue data
tesla_revenue = tesla.quarterly_financials.loc['Total Revenue']
tesla_revenue = pd.DataFrame(tesla_revenue).reset_index()
tesla_revenue.columns = ['Date', 'Revenue']

# Sort by date
tesla_revenue = tesla_revenue.sort_values('Date')
print(tesla_revenue.tail())
```

# Question 3: Extracting GameStop Stock Data Using yfinance

```python
# Download GameStop data
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")

# Reset index
gme_data.reset_index(inplace=True)

# Display first 5 rows
print(gme_data.head())
```

```python
# Get quarterly revenue data
gme_revenue = gme.quarterly_financials.loc['Total Revenue']
gme_revenue = pd.DataFrame(gme_revenue).reset_index()
gme_revenue.columns = ['Date', 'Revenue']

# Sort by date
gme_revenue = gme_revenue.sort_values('Date')
print(gme_revenue.tail())
```

```python
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'], data['Close'])
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price ($)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

make_graph(tesla_data, 'Tesla Stock Price History')
```

```python
make_graph(gme_data, 'GameStop Stock Price History')
```
